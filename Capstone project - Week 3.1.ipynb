```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
bsoup = BeautifulSoup(url,'lxml')
```

```python
#print(bsoup.prettify())
```

```python
table_post = bsoup.find('table') # find the table in the html
fields = table_post.find_all('td') # contains the elements in the table

# make lists to contain the elements of the columns of our dataframe
postcode = []
borough = []
neighbourhoods = []

#populate the dataframe
for i in range(0, len(fields)):
    # the table elements are links elements so we have to find 'a' in the html
    ''' example : [<a href="/wiki/Downtown_Toronto" title="Downtown Toronto">Downtown Toronto</a>,
 <a href="/wiki/Regent_Park" title="Regent Park">Regent Park</a>,
 <a class="mw-redirect" href="/wiki/Harbourfront,_Toronto" title="Harbourfront, Toronto">Harbourfront</a>]'''
    
    contents = fields[i].find_all('a')
    if len(contents) == 0:
        continue # case of Not assigned borough
    else:
        borough.append(contents[0].text.strip())
        
        if(len(contents)>1):
            neighbourhood = contents[1].text.strip()
            for n in range(2,len(contents)): # more neighbourhoods
                neighbourhood += ', ' + contents[n].text.strip()
            neighbourhoods.append(neighbourhood)
        else:
            neighbourhoods.append(contents[0].text.strip())
    postcode.append(fields[i].b.text.strip())
    
# put in pandas dataframe
df_pc = pd.DataFrame(data=[postcode, borough, neighbourhoods]).transpose()
df_pc.columns = ['Postcode', 'Borough', 'Neighbourhood']
```
